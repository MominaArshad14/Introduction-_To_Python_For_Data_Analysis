```python
df_csv = pd.read_csv("Book 12(Sheet1).csv")
df_csv.head()
```

```python
df = df_csv   # use  loaded data

# First five rows
df.head()

# Summary of dataset
df.info()

# Basic statistics
df.describe()

# Columns names
df.columns
```

```python
mean_value = df["Y"].mean()
median_value = df["Y"].median()
std_value = df["Y"].std()
min_value = df["Y"].min()
max_value = df["Y"].max()

mean_value, median_value, std_value, min_value, max_value
```

```python
df.corr()
```

```python
#Data preparation
# Drop rows where X1 or Y are NaN
df_clean = df[["X1", "Y"]].dropna()
# Prepare data
X = df_clean[["X1"]]   # 2D dataframe
y = df_clean["Y"]      # 1D series

```

```python
#Model Selection
model = LinearRegression()
model.fit(X, y)
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
print("R² Score:", model.score(X, y))
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("X1")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.show()
```
